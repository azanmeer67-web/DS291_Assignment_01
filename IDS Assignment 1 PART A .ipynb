**Q1. Arithmetic and Assignment Operators**

```python
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)

a += 5
print("After a += 5:", a)

a -= 3
print("After a -= 3:", a)

a *= 2
print("After a *= 2:", a)

a /= 4
print("After a /= 4:", a)
```

**Q2. Comparison and Logical Operators**

```python
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

print("x == y:", x == y)
print("x != y:", x != y)
print("x > y:", x > y)
print("x < y:", x < y)
print("x >= y:", x >= y)
print("x <= y:", x <= y)

print("x > y AND x != 0:", (x > y and x != 0))
print("x < y OR y == 0:", (x < y or y == 0))
print("NOT(x == y):", not(x == y))
```

**Q3. Bitwise Operators**

```python
p = int(input("Enter first integer: "))
q = int(input("Enter second integer: "))

print("Bitwise AND:", p & q)
print("Bitwise OR:", p | q)
print("Bitwise XOR:", p ^ q)
print("Bitwise NOT (~p):", ~p)
print("Left Shift p << 2:", p << 2)
print("Right Shift p >> 2:", p >> 2)
```

Q4. Identity and Membership **Operators**

```python
numbers = [10, 20, 30, 40, 50]

a = numbers
b = numbers.copy()

print("a is b:", a is b)
print("a is not b:", a is not b)

print("30 in list:", 30 in numbers)
print("100 not in list:", 100 not in numbers)
```

**Q5. String Operations**

```python
text = "Data science is an exciting field."

print("Length:", len(text))
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())


print("Is 'science' in text?", "science" in text)


print("Replace 'exciting' with 'growing':", text.replace("exciting", "growing"))

print("Sliced text (0 to 10):", text[0:10])

print("Reversed string:", text[::-1])
```
